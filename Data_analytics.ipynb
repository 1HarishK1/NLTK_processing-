**Concatenating I'd and URL**

```python
title=[]
for i in range(0,len(data["URL"])):
       title.append(f"{data['URL'][i]}{int(data['URL_ID'][i])}")
title
```

**Scraping page**

```python


i=0
topic=[]
while i<114:
  response=re.get(title[i])
  soup=BeautifulSoup(response.text,"html.parser")
  name=soup.find('div',class_='td-container')
  if name is  None:
    topic.append("Page is error")
  else:
   name=name.find('header',class_="td-post-title")
   name=name.find('h1',class_="entry-title")
   topic.append(name.text)
  i+=1
```

**Writing contents from txt files and Concate as whole for Using as Stopwords**

```python
Stopwords=['StopWords_Auditor.txt',
          'StopWords_Generic.txt',
          'StopWords_DatesandNumbers.txt',
          'StopWords_Geographic.txt',
          'StopWords_Names.txt']
```

```python

all_stopwords=" "
for txts in Stopwords:
   file=open(txts,'r',encoding="ISO-8859-1")
   content=file.read()
   content=content.lower()
   all_stopwords+=content
   file.close()
print(all_stopwords)
```

```python
H=word_tokenize(topic[1])
print(H[3])
for i in H:
   print(i)
```

**Preprocessing the title and store in new list**

```python

hea=[]
def token_and_remove(char):
    global all_stopwords
    H=word_tokenize(char)
    new_char=""
    for sent in H:
      if sent not in all_stopwords:
        new_char=new_char+' '+sent
    hea.append(new_char.split())
    return hea
```

```python

for senc in range(0,a):
  token_and_remove(topic[senc])
```

**calculating avg words  , total words**

```python
per_sent=[]
tot_words=0
for i in hea:
   wrd_sent=0
   for j in i:
      wrd_sent+=1
      tot_words+=1
   per_sent.append(wrd_sent)
print(tot_words)
```

**Polarity_score calculation Using Vader**

```python
sentiment=SentimentIntensityAnalyzer()
```

```python
dic=[]
trt=[]
```

```python
for senc in range(0,(len(hea))):
   value=sentiment.polarity_scores(topic[senc])
   new=sentiment.polarity_scores(hea[senc])
  #calculating for original cleaned data
   dic.append(value)
  #calculating for preprocessed original cleaned data
   trt.append(new)
```

```python
ps=[]
neg=[]
for i in range (len(trt)):
    ps.append(round(trt[i]["pos"]*100))
    neg.append(round(trt[i]["neg"]*100))
```

```python
avg_sent=[]
for i in per_sent:
   avg_sent.append(tot_words//len(topic))
```

**Calculations**

```python
avg_ch_len=[]
for ch in per_sent:
   avg_ch_len.append(Total//ch)
```

```python
polar=[]
sub=[]
for i in range(len(ps)):
   cal=((ps[i]+neg[i])//tot_words)
   sub.append(cal)
   if (ps[i]!= 0.0) or (neg[i]!=0.0):
     value=((ps[i]-neg[i]) //(ps[i]+neg[i]))
     polar.append(value)
   else:
     value=0
     polar.append(value)
```

**Storing the calculated data  in Excel**

```python
data=pd.DataFrame({'POSITIVE_SCORE':ps})
data=pd.DataFrame({'NEGATIVE_SCORE':neg})
data=pd.DataFrame({'POLARITY_SCORE':polar})
data=pd.DataFrame({'SUBJECTIVITY_SCORE':sub})
data=pd.DataFrame({'WORD_COUNT':per_sent})
data=pd.DataFrame({'AVG_NO_OF_WRDS':avg_sent})
data=pd.DataFrame({'AVG_COUNT_LENGTH':avg_ch_len})
```
